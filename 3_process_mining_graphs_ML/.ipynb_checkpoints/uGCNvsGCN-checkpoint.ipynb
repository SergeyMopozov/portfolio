# Experiment design

* Load 6 datasets from PyG TU collection (2 variants: original and cleaned)
* 12 rounds of random splits by PyG for every dataset
* report best (B) and average (A) accuracy for GCN
* compare against simple baselines: degree kernel, noisy graph convolution, and a combination of both.

```python
from torch.nn import Linear
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.nn import global_mean_pool


class GCN(torch.nn.Module):
    def __init__(self, hidden_channels):
        super(GCN, self).__init__()
        torch.manual_seed(myRandomSeed)
        self.conv1 = GCNConv(dataset.num_node_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.conv3 = GCNConv(hidden_channels, hidden_channels)
        self.lin = Linear(hidden_channels, dataset.num_classes)

    def forward(self, x, edge_index, batch):
        # 1. Obtain node embeddings 
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = self.conv2(x, edge_index)
        x = x.relu()
        x = self.conv3(x, edge_index)

        # 2. Readout layer
        x = global_mean_pool(x, batch)  # [batch_size, hidden_channels]

        # 3. Apply a final classifier
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.lin(x)
        
        return x

model = GCN(hidden_channels=64)
print(model)
```

## Helper functions

```python
def make_GCN_weights(dataset, hidden_channels, seed):
    np.random.seed(seed=seed)

    N = dataset.num_features
    M = hidden_channels

    W0 = np.random.rand(N, M)*2 - 1
    W1 = np.random.rand(N + M, M)*2 - 1
    W2 = np.random.rand(N + 2*M, M)*2 - 1

    return W0, W1, W2

W0, W1, W2 = make_GCN_weights(dataset, 64, myRandomSeed)

print (W0.shape)
print (W1.shape)
print (W2.shape)
```

```python
def make_stats(datum, W0, W1, W2):
    g = to_networkx(datum)
    numNodes = len(g.nodes)
    # degree kernel
    degreeHist = nx.degree_histogram(g)
    # normalize
    degreeHist = [x/numNodes for x in degreeHist]
    maxDegree = len(degreeHist)
    # graph label
    y = int(datum['y'])
    # let's make graph embeddings
    X = np.array(datum['x'])
    A = nx.convert_matrix.to_scipy_sparse_matrix(g)
    D = (csgraph.laplacian(A, normed=True))
    # layer 0
    Xc = D @ X @ W0
    # ReLU
    Xc = Xc * (Xc>0)
    Xn = np.hstack((X, Xc))
    # layer 1
    Xc = D @ Xn @ W1
    # ReLU
    Xc = Xc * (Xc>0)
    Xn = np.hstack((Xn, Xc))
    # layer 2 - node embeddings
    Xc = D @ Xn @ W2
    # mean pooling - graph embedding
    embedding = Xc.sum(axis=0) / Xc.shape[0]
    
    return [maxDegree, degreeHist, embedding, y]

def wrap_stats(dataset, W0, W1, W2, limit = 10):
    i = 0
    statsList = []
    for datum in dataset:
        statsList.append(make_stats(datum, W0, W1, W2))
        #print(i)
        i += 1
        if i == limit:
            break
    
    myColumns = ['maxDegree', 'degreeHist', 'embedding', 'y']
    output = pd.DataFrame(statsList, columns=myColumns)
    
    return output
```

## Tournament

Beware! Runnig this cell takes ~4 hours

```python
%%time
datasetNames = ['AIDS', 'BZR', 'COX2', 'DHFR', 'MUTAG', 'PROTEINS']
numRounds = 12
numEpochs = 200

# store counts for DK, uGCN, Mix, and a trained GCN wins over rounds
counterA = [0,0,0,0]
counterB = [0,0,0,0]
contenders = ['DK', 'uGCN', 'Mix', 'GCN']
awardHeaders = contenders + ['dataset', 'cleaned', 'round', 'AB']
statisticsHeaders = ['dataset', 'cleaned', 'round', 'model', 'testAccuracy']

statistics = []
awards = []

for name in datasetNames:

    for cleaned in ['yes', 'no']:

        if cleaned == 'yes':
            dataset = TUDataset(root='data/TUDatasetC', name=name, cleaned=True)
        else:
            dataset = TUDataset(root='data/TUDataset', name=name, cleaned=False)

        counterDatasetA = [0,0,0,0]
        counterDatasetB = [0,0,0,0]

        for round in range(numRounds):
            counterRoundA = [0,0,0,0, name, cleaned, round, 'A']
            counterRoundB = [0,0,0,0, name, cleaned, round, 'B']
            roundResultsA = []
            roundResultsB = []

            torch.manual_seed(myRandomSeed+round)
            dataset = dataset.shuffle()

            class GCN(torch.nn.Module):
                def __init__(self, hidden_channels):
                    super(GCN, self).__init__()
                    torch.manual_seed(myRandomSeed+round)
                    self.conv1 = GCNConv(dataset.num_node_features, hidden_channels)
                    self.conv2 = GCNConv(hidden_channels, hidden_channels)
                    self.conv3 = GCNConv(hidden_channels, hidden_channels)
                    self.lin = Linear(hidden_channels, dataset.num_classes)

                def forward(self, x, edge_index, batch):
                    # 1. Obtain node embeddings 
                    x = self.conv1(x, edge_index)
                    x = x.relu()
                    x = self.conv2(x, edge_index)
                    x = x.relu()
                    x = self.conv3(x, edge_index)

                    # 2. Readout layer
                    x = global_mean_pool(x, batch)  # [batch_size, hidden_channels]

                    # 3. Apply a final classifier
                    x = F.dropout(x, p=0.5, training=self.training)
                    x = self.lin(x)
                    
                    return x

            model = GCN(hidden_channels=64)        

            # 80/20 split
            treshold = int(len(dataset)*.8)

            trainDataset = dataset[:treshold]
            testDataset = dataset[treshold:]

            W0, W1, W2 = make_GCN_weights(dataset, 64, myRandomSeed)

            dfTrain = wrap_stats(trainDataset, W0, W1, W2, limit = 10000)
            dfTest = wrap_stats(testDataset, W0, W1, W2, limit = 10000)

            maxDegree = max(dfTrain['maxDegree'].max(), dfTest['maxDegree'].max())

            def expander (featureList, dimension=maxDegree):
                delta = dimension - len(featureList)
                for i in range(delta):
                    featureList.append(0)
                return featureList

            dfTrain['degreeHist'] = dfTrain['degreeHist'].apply(expander)
            dfTest['degreeHist'] = dfTest['degreeHist'].apply(expander)

            myRandomSeedNow = myRandomSeed + round
            print()
            print(f'Dataset: {dataset}:')
            print('====================')
            print(f'Number of graphs: {len(dataset)}')
            print(f'Random seed: {myRandomSeedNow}')
            print(f'Number of training graphs: {len(trainDataset)}')
            print(f'Number of test graphs: {len(testDataset)}')
            print(f'Maximum degree: {maxDegree}')
            print('=============================================================')
            print(f'Round {round}')
            print('=============================================================')

            # degree kernel
            print('degree kernel')
            DK_train = pd.DataFrame([{x: y for x, y in enumerate(item)} for item in dfTrain['degreeHist'].values.tolist()], index=dfTrain.index)
            X_train = DK_train.values
            y_train = dfTrain['y'].values
            DK_test = pd.DataFrame([{x: y for x, y in enumerate(item)} for item in dfTest['degreeHist'].values.tolist()], index=dfTest.index)
            X_test = DK_test.values
            y_test = dfTest['y'].values

            testAccuracy = random_forest() 
            roundResultsA.append(float(testAccuracy))
            roundResultsB.append(testAccuracy)

            roundStatistics = [name, cleaned, round, 'DK', testAccuracy]

            statistics.append(roundStatistics)

            # unsupervised GCN
            print('unsupervised GCN')
            U_train = pd.DataFrame([{x: y for x, y in enumerate(item)} for item in dfTrain['embedding'].values.tolist()], index=dfTrain.index)
            X_train = U_train.values
            U_test = pd.DataFrame([{x: y for x, y in enumerate(item)} for item in dfTest['embedding'].values.tolist()], index=dfTest.index)
            X_test = U_test.values
            
            testAccuracy = random_forest() 
            roundResultsA.append(float(testAccuracy))
            roundResultsB.append(testAccuracy)

            roundStatistics = [name, cleaned, round, 'uGCN', testAccuracy]

            statistics.append(roundStatistics)

            # Both 
            print('degree kernel + unsupervised GCN')
            U_train.columns = ['u'+str(x) for x in U_train.columns]
            U_test.columns = ['u'+str(x) for x in U_test.columns]

            DKU_train = DK_train.join(U_train)
            DKU_test = DK_test.join(U_test)

            X_train = DKU_train.values
            X_test = DKU_test.values

            testAccuracy = random_forest() 
            roundResultsA.append(float(testAccuracy))
            roundResultsB.append(testAccuracy)

            roundStatistics = [name, cleaned, round, 'Mix', testAccuracy]
            statistics.append(roundStatistics)

            # TRAIN GCN HERE
            testAccuracyAverage = 0
            testAccuracyBest = 0 
            testAccuracyList = []

            # prepare data loaders
            train_loader = DataLoader(trainDataset, batch_size=64, shuffle=True)
            test_loader = DataLoader(testDataset, batch_size=64, shuffle=False)

            # train the model

            optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
            criterion = torch.nn.CrossEntropyLoss()

            def train():
                model.train()

                for data in train_loader:  # Iterate in batches over the training dataset.
                    out = model(data.x, data.edge_index, data.batch)  # Perform a single forward pass.
                    loss = criterion(out, data.y)  # Compute the loss.
                    loss.backward()  # Derive gradients.
                    optimizer.step()  # Update parameters based on gradients.
                    optimizer.zero_grad()  # Clear gradients.

            def test(loader):
                model.eval()

                correct = 0
                for data in loader:  # Iterate in batches over the training/test dataset.
                    out = model(data.x, data.edge_index, data.batch)  
                    pred = out.argmax(dim=1)  # Use the class with highest probability.
                    correct += int((pred == data.y).sum())  # Check against ground-truth labels.
                return correct / len(loader.dataset)  # Derive ratio of correct predictions.


            for epoch in range(1, numEpochs + 1):
                train()
                train_acc = test(train_loader)
                test_acc = test(test_loader)
                #print(f'Epoch: {epoch:03d}, Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}')

                testAccuracyList.append(test_acc)
                if test_acc > testAccuracyBest:
                    testAccuracyBest = test_acc


            testAccuracyAverage = np.mean(testAccuracyList[-10:])

            roundResultsA.append(testAccuracyAverage)
            roundResultsB.append(testAccuracyBest)

            roundStatisticsA = [name, cleaned, round, 'GCN-A', testAccuracyAverage]
            roundStatisticsB = [name, cleaned, round, 'GCN-B', testAccuracyBest]

            statistics.append(roundStatisticsA)
            statistics.append(roundStatisticsB)

            print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
            print('~~~~~~~~~~~~~ TRAINED GCN ~~~~~~~~~~~~')
            print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
            print(f'Best accuracy over {epoch} epochs: {testAccuracyBest}')
            print(f'Average accuracy over last 10 epochs: {testAccuracyAverage}')
            print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

            # END GCN TRAINING

            counterDatasetA[np.argmax(roundResultsA)] +=1
            counterA[np.argmax(roundResultsA)] +=1
            counterRoundA[np.argmax(roundResultsA)] +=1
            awards.append(counterRoundA)

            counterDatasetB[np.argmax(roundResultsB)] +=1
            counterB[np.argmax(roundResultsB)] +=1
            counterRoundB[np.argmax(roundResultsB)] +=1
            awards.append(counterRoundB)

            print('=============================================================')
            print('=============================================================')
        
        print('*******************')
        print('*******************')
        print('*******************')
        print('*******************')
        print('*******************')

        roundNo = round + 1
        print('Setting A')
        print(f'Results for the dataset over {roundNo} rounds')

        print(counterDatasetA)
        champ = contenders[np.argmax(counterDatasetA)]
        print(f'Dataset {name} Champion: {champ}')
        
        print('Cumulative wins over all datasets:')
        print(counterA)

        print('*******************')

        print('Setting B')
        print(f'Results for the dataset over {roundNo} rounds')

        print(counterDatasetB)
        champ = contenders[np.argmax(counterDatasetB)]
        print(f'Dataset Champion: {champ}')
        print('Cumulative wins over all datasets:')
        print(counterB)

        print('*******************')
        print('*******************')
        print('*******************')
        print('*******************')
        print('*******************')
        print('*******************')

statisticsDF = pd.DataFrame(statistics, columns=statisticsHeaders)
awardsDF = pd.DataFrame(awards, columns=awardHeaders)
```
