## 1. Ресемплинг

```python
min_day = data['day'].min()
max_day = data['day'].max()

print(min_day)
print(max_day)
```

```python
data['id'].nunique()
```

```python
resample_data = data.pivot_table(index='id', 
                                 columns='day', 
                                 values='value_1', 
                                 aggfunc='sum', 
                                 fill_value=0)\
                    .reset_index()\
                    .melt(id_vars='id')

category_dict = data[category_cols].drop_duplicates().sort_values('id').reset_index(drop=True)
value_1_data = pd.merge(category_dict, resample_data,  on='id', how='right')

#проверим корректность ресемлпинга
assert value_1_data.shape[0] == len(pd.date_range(min_day, max_day)) * data['id'].nunique()
```

## Feature exctraction¶

The model needs features, and all we have is a 1-dimentional time series. What features can we exctract?
1. Features adds by processed row data:
    Lags and diffs of time series
    
    Window statistics:
        Max/min value of series in a window
        Average/median value in a window
        Window variance
        etc.
    Forecasts from other models (note that we can lose the speed of prediction this way)
        
2. Features merged on date:        
    Date and time features:
        Minute of an hour, hour of a day, day of the week, and so on
        Is this day a holiday? Maybe there is a special event? Represent that as a boolean feature

3. Categorical features        
    Target encoding
    
    

## 1. Тренд и сезонность.

```python
idx = 'id'
value_1_data['trend'] = 1
value_1_data['trend'] = value_1_data.groupby(idx)['trend'].cumsum()
value_1_data['trend_log'] = value_1_data.groupby(idx)['trend'].apply(np.log1p)
```

```python
# признаки для недельной сезонности

for K in range(1,3):
    value_1_data['sin_week_season_' + str(K)] = np.sin(value_1_data['trend'] * 2 * np.pi * K /7)
    value_1_data['cos_week_season_' + str(K)] = np.cos(value_1_data['trend'] * 2 * np.pi * K /7)
    
# признаки для годовой сезонности
for K in range(1,3):
    value_1_data['sin_year_season_' + str(K)] = np.sin(value_1_data['trend'] * 2 * np.pi * K /365.25)
    value_1_data['cos_year_season_' + str(K)] = np.cos(value_1_data['trend'] * 2 * np.pi * K /365.25)
```

## 2. Кодирование категориальных переменных

```python
encoders = []
for col in category_cols:
    enc = LabelEncoder()
    value_1_data[col] = enc.fit_transform(value_1_data[col])
    encoders.append(enc)
```

```python
def plot_ts(data, idx, x, y):
    data = data[data['id']==idx]
    plt.figure(figsize=(16,6))
    sns.lineplot(x, y, data=data)
    plt.grid()
    plt.show()
```

## 3. Календарные фичи

```python
calendar = value_1_data[value_1_data['id']==0][['day']].copy()
```

```python
calendar['year'] = calendar['day'].dt.year
calendar['quarter'] = calendar['day'].dt.quarter
calendar['month'] = calendar['day'].dt.month
calendar['week'] = calendar['day'].dt.week
calendar['day_num'] = calendar['day'].dt.day
calendar['dayofweek'] = calendar['day'].dt.dayofweek

def get_week_ends(dayofweek):
    if dayofweek == 5 or dayofweek == 6:
        return 1
    else:
        return 0
    
calendar['weekends'] = calendar['dayofweek'].apply(get_week_ends)

# add calendar
value_1_data = pd.merge(value_1_data, calendar, on='day')
```

## 4. Лаговые фичи

```python
# lags of series
lag_start = 1
lag_end = 7

lag_list = list(range(lag_start, lag_end))
lag_list.append(14)
lag_list.append(21)
lag_list.append(28)

print(lag_list)
idx = 'id'
target = 'value'
for i in lag_list:
        value_1_data["lag_{}".format(i)] = value_1_data.groupby(idx)[target].shift(i)
```

## 5. Прирост

```python
dif_lag_start = 1
dif_lag_end = 7

lag_list = list(range(dif_lag_start, dif_lag_end))
lag_list.append(14)
lag_list.append(21)
lag_list.append(28)

print(lag_list)

idx = 'id'
target = 'value'
for i in lag_list:
        value_1_data["dif_{}".format(i)] = value_1_data.groupby(idx)[target].diff(i)
```

## 6. Скользящие статистики

```python
# скользящее окно
window_size= [7, 14, 28]
predict_periods = 1
idx = 'id'
target = 'value'
for i in window_size:
        value_1_data["rol_mean_{}".format(i)] = value_1_data.groupby(idx)[target]\
                                                        .shift(predict_periods)\
                                                        .rolling(i)\
                                                        .mean()
        value_1_data["rol_min_{}".format(i)] = value_1_data.groupby(idx)[target]\
                                                        .shift(predict_periods)\
                                                        .rolling(i)\
                                                        .min()
        
        value_1_data["rol_max_{}".format(i)] = value_1_data.groupby(idx)[target]\
                                                        .shift(predict_periods)\
                                                        .rolling(i)\
                                                        .max()
        value_1_data["rol_var_{}".format(i)] = value_1_data.groupby(idx)[target]\
                                                        .shift(predict_periods)\
                                                        .rolling(i)\
                                                        .var()
```

## 8. Target encoding

```python
def code_mean(data, cat_feature, real_feature):
    """
    Returns a dictionary where keys are unique categories of the cat_feature,
    and values are means over real_feature
    """
    return dict(data.groupby(['id', cat_feature])[real_feature].mean())

def get_code_mean(data, target, category_list, test_size):
    test_index = int(len(data.dropna())*(1-test_size))
    for cat in category_list:
        data['average' + f'_{cat}'] = list(map(code_mean(data[:test_index], cat, target).get, 
                                               zip(data['id'], data[cat])))
    return data
```

## 7. "Дамми"-переменные

```python
dummies_cols = ['group_0', 'group_1', 'group_2', 'group_3']
dummies_data = value_1_data.copy()
for col in dummies_cols:
    dummies = pd.get_dummies(dummies_data[col], drop_first=True)
    dummies.columns = [col + '_' + str(dc) for dc in dummies.columns]
    dummies_data = pd.concat([dummies_data, dummies], axis=1)

    
value_1_data = dummies_data.drop(columns=dummies_cols)
```
